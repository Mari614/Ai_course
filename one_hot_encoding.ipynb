```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
car=pd.read_csv('carprices.csv')
car
```

```python
x= car.drop(columns=['Sell Price($)' ])
y= car['Sell Price($)']
```

```python
%matplotlib inline
plt.xlabel( 'Mileage' , 'Age(yrs)') 
plt.ylabel('Sell Price($)')
plt.scatter(car[Age(yrs) ,  Mileage] , car['Sell Price($)'],color='red',marker='*')
```

```python
dummies=pd.get_dummies(car['Car Model'])
dummies
```

```python
merged=pd.concat([car, dummies] , axis='columns')
merged
```

```python
final=merged.drop(['Car Model' , 'Mercedez Benz C class'], axis='columns')
final
```

```python
x= final.drop(columns=['Sell Price($)' ])
y= final['Sell Price($)']
x
```

```python
model=linear_model.LinearRegression()
model.fit(x ,y)
```

```python
model.predict([[45000,4,0,0]])
```

```python
model.predict([[86000,7,0,1]])
```

```python
model.score(x ,y)
```
